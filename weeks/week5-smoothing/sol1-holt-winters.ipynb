# Problem 1 `Setting a DatetimeIndex Frequency`
What do you notice about the DatetimeIndex's frequency? In order to build a Holt-Winters smoothing model, statsmodels needs to know the frequency of the data (whether it's daily, monthly etc.). Our observations occur at the start of each month, use the right alies from [this list](http://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases) to set the index's frequency

```python
df.index.freq = 'MS'
df.index
```

# Problem 2 `Simple Exponential Smoothing`

We compute the weighted moving average using the `alpha` below & then store the result in the `EWMA12` column.  


Apply `SimpleExpSmoothing` (with the `alpha` below) on the 'Thousands of Passengers' column and then show how results are the same as the  the weighted moving average as the pandas <tt>.ewm()</tt> method:<br>

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

span = 12
alpha = 2/(span+1)

df['EWMA12'] = df['Thousands of Passengers'].ewm(alpha=alpha,adjust=False).mean()
```

```python
df['SES12']=SimpleExpSmoothing(df['Thousands of Passengers']).fit(smoothing_level=alpha,optimized=False).fittedvalues.shift(-1)
df.head()
```

<div class="alert alert-danger"><strong>NOTE:</strong> For some reason, when <tt>optimized=False</tt> is passed into <tt>.fit()</tt>, the statsmodels <tt>SimpleExpSmoothing</tt> function shifts fitted values down one row. We fix this by adding <tt>.shift(-1)</tt> after <tt>.fittedvalues</tt></div>

___
# Problem 3 `Double Exponential Smoothing`
Where Simple Exponential Smoothing employs just one smoothing factor $\alpha$ (alpha), Double Exponential Smoothing adds a second smoothing factor $\beta$ (beta) that addresses trends in the data. Like the alpha factor, values for the beta factor fall between zero and one ($0<\betaâ‰¤1$). The benefit here is that the model can anticipate future increases or decreases where the level model would only work from recent calculations.

Fit an addititive `ExponentialSmoothing` to the 'Thousands of Passengers' column & store the results back in a new column in the df


```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

df['DESadd12'] = ExponentialSmoothing(df['Thousands of Passengers'], trend='add').fit().fittedvalues.shift(-1)
df.head()
```

# Problem 4 `Plot both smoothings with the original series`
Plot all both smoothings with the original series & then determine which of the 2 smoothings visually has the better fit

```python
df[['Thousands of Passengers','EWMA12','DESadd12']].iloc[:24].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

# Problem 5 `Multiplicative Exponential Smoothing` 
Fit an multiplicative `ExponentialSmoothing` to the 'Thousands of Passengers' column & store the results back in a new column in the df.  Does multiplicative give better results than the additive version?

```python
df['DESmul12'] = ExponentialSmoothing(df['Thousands of Passengers'], trend='mul').fit().fittedvalues.shift(-1)
df.head()
```

```python
df[['Thousands of Passengers','DESadd12','DESmul12']].iloc[:24].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

___
# Problem 6 `Triple Exponential Smoothing`
Triple Exponential Smoothing, the method most closely associated with Holt-Winters, adds support for both trends and seasonality in the data. 

Fit both an additivite and a multiplicative Triple Exponential Smoothing to our series & store the fitted values in 2 new columns

```python
df['TESadd12'] = ExponentialSmoothing(df['Thousands of Passengers'],trend='add',seasonal='add',seasonal_periods=12).fit().fittedvalues
df['TESmul12'] = ExponentialSmoothing(df['Thousands of Passengers'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
df.head()
```

Plot fitted values of both triple exponential smoothing
