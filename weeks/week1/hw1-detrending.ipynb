# Detrending Techniques
In this assignment, you will detrend times series using various techniques such as
* Differencing
* SciPy's Signal
* HP Filter

## Let's load a data set
In this lesson, we will use the Minimum Daily Temperatures dataset as an example. This dataset describes the minimum daily temperatures over 10 years (1981-1990) in the city Melbourne, Australia

```python
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('../../data/daily-min-temperatures.csv', header=0, index_col=0,
    parse_dates=True, squeeze=True)
pyplot.figure(figsize=(15,6))
pyplot.plot(series)
pyplot.xlabel('Date')
pyplot.ylabel('Temp')
pyplot.title('Min Daily Temp', fontsize=16)
pyplot.show()
```

## Detrend using Differencing
Detrend this time series using the differencing technique & then plot the result to visually if how the seasonality was removed

```python
diff = series.diff()
pyplot.figure(figsize=(15,6))
pyplot.plot(diff)
pyplot.title('Detrending using Differencing', fontsize=16)
pyplot.xlabel('Date')
pyplot.ylabel('Temp')
pyplot.show()
```

## Load Airline Passengers DataSet
Let's load the airline-passengers dataset and plot it to observe both the seasonal and linear trend within it 

```python
series = read_csv('../../data/airline-passengers.csv', header=0, index_col=0, parse_dates=True,
    squeeze=True)
pyplot.figure(figsize=(15,6))
pyplot.plot(series)
pyplot.show()
```

## Detrend using SciPy Signal
Now apply SciPy's Signal method to remove the linear trend.  Plot the detrended series & explain visually why the linear trend is now long present.

```python
from scipy import signal
detrended = signal.detrend(series)
pyplot.figure(figsize=(15,6))
pyplot.plot(detrended)
pyplot.title('Detrending using SciPy Signal', fontsize=16)
pyplot.xlabel('Date')
pyplot.ylabel('Passengers')
pyplot.show()
```

## Detrend using HP Filter
Now apply HP Filter to detect & remove the linear trend

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
cycle,trend = hpfilter(series, lamb=1600)
detrended = series - trend
pyplot.figure(figsize=(15,6))
pyplot.plot(detrended)
pyplot.title('Detrending using HP Filter', fontsize=16)
pyplot.xlabel('Date')
pyplot.ylabel('Passengers')
pyplot.show()
```
