## Imports

```python
from pandas import read_csv
from matplotlib import pyplot
import numpy as np
from numpy import asarray
from pandas import DataFrame
from pandas import concat
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor 
```

## Data

```python
bikes = read_csv('Bike_Sharing_Demand.csv', header=0)[:1000]
values = []
for i in range(bikes.shape[0]):
    values.append(np.array([bikes.iloc[i]['count']]))
values = np.asarray(values)    
# plot dataset
pyplot.plot(values)
pyplot.show()
```

# Problem 1: `Transform the data from a series to a supervised dataset`
The RF classifier will require a 2d numpy array input and an array of target values. To create forecast model, for each realized datetime record of demand, you should generate 6 additional columns which record the 1-6 periods prior value (i.e. the demand 1 hr, 2hrs ago, etc). Be sure to drop any nan values from the datafram before trying to cast back to a 2d array (which cannot have nan as inputs to RF) .

```python
def to_supervised(data, n_in=1, n_out=1, dropnan=True):
	n_vars = 1 if type(data) is list else data.shape[1]
	df = DataFrame(data)
	cols = list()
	# input sequence (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df.shift(i))
	# forecast sequence (t, t+1, ... t+n)
	for i in range(0, n_out):
		cols.append(df.shift(-i))
	# put it all together
	agg = concat(cols, axis=1)
	# drop rows with NaN values
	if dropnan:
		agg.dropna(inplace=True)
	return agg.values
```

```python
data = to_supervised(values, n_in=6)
```

# Problem 2: `Perform Train/Test Split`
Train on all but the last 72 hours of the dataset provided.

```python
# split a univariate dataset into train/test sets
def train_test_split(data, n_test):
	return data[:-n_test, :], data[-n_test:, :]
```

# Problem 3: `Define RF classifer to use 1000 iterations`
Define the model using RandomForestRegressor from scikit.

```python
def random_forest_forecast(train, testX):
	# transform list into array
	train = asarray(train)
	# split into input and output columns
	trainX, trainy = train[:, :-1], train[:, -1]
	# fit model
	model = RandomForestRegressor(n_estimators=1000)
	model.fit(trainX, trainy)
	# make a one-step prediction
	yhat = model.predict([testX])
	return yhat[0]
```

# Problem 4: `Define walk-foward validation of test set and measure MAE (Mean Absolute Error)`

Create a function or loop to define errors measured on test set between actual and predicted hourly bike sharing demand.

```python
# walk-forward validation for univariate data
def walk_forward_validation(data, n_test):
	predictions = list()
	# split dataset
	train, test = train_test_split(data, n_test)
	# seed history with training dataset
	history = [x for x in train]
	# step over each time-step in the test set
	for i in range(len(test)):
		# split test row into input and output columns
		testX, testy = test[i, :-1], test[i, -1]
		# fit model on history and make a prediction
		yhat = random_forest_forecast(history, testX)
		# store forecast in list of predictions
		predictions.append(yhat)
		# add actual observation to history for the next loop
		history.append(test[i])
		# summarize progress
		print('>expected=%.1f, predicted=%.1f' % (testy, yhat))
	# estimate prediction error
	error = mean_absolute_error(test[:, -1], predictions)
	return error, test[:, -1], predictions
```

# Problem 5: `Fit model and visualize expected vs predicted values`

Use the validation tool you created in the prior step and report the MAE and show a plot of actual vs predicted.

```python
# evaluate
mae, y, yhat = walk_forward_validation(data, 72)
print('MAE: %.3f' % mae)
# plot expected vs predicted
pyplot.plot(y, label='Expected')
pyplot.plot(yhat, label='Predicted')
pyplot.legend()
pyplot.show()
```
