# Homework 1 `Various Types of TS plots`
For this assignment, we'll be working with the daily minimum temperatures over 10 years (1981-1990) in the city of Melbourne, Austrailia
Let's load that into a pandas dataframe

```python
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('../../data/daily-minimum-temperatures.csv', header=0, index_col=0,
    parse_dates=True, squeeze=True)
series.head()
```

# Problem 1 `Line Plot`
Render a basic line plot of this series

```python
series.plot()
pyplot.show()
```

# Problem 2 `Dot Plot`
Create a basic dot plot

```python
series.plot(style='k.')
pyplot.show()
```

# Problem 3 `Stacked Line Plots`
First group series by year and then create a stacked line plot

```python
from pandas import Grouper
from pandas import DataFrame
groups = series.groupby(Grouper(freq='A')) 
years = DataFrame()
for name, group in groups:
  years[name.year] = group.values
years.plot(subplots=True, legend=False)
pyplot.show()
```

# Problem 4 `Histogram`
Create a histogram of the series

```python
series.hist();
```

# Problem 5 `Density Plot`

```python
series.plot(kind='kde');
```

# Problem 6 `Box Plot`
Create a box plot of the series grouped by year = `years`

```python
years.boxplot();
```

# Problem 7 `Heat Map`
Create a heat map of the series grouped by year (i.e. `years`)

```python
years = years.T
pyplot.matshow(years, interpolation=None, aspect='auto') 
pyplot.show()
```

# Problem 8 `Lag Plot`
Create a lag plot of the series

```python
from pandas.plotting import lag_plot
lag_plot(series)
pyplot.show()
```

# Problem 9 `Multiple Scatter Plots` 
Create a scatter plot for the observation with each value in the previous seven days.

```python
from pandas import concat
values = DataFrame(series.values)
lags = 7
columns = [values]
for i in range(1,(lags + 1)):
	columns.append(values.shift(i))
dataframe = concat(columns, axis=1)
columns = ['t']
for i in range(1,(lags + 1)):
	columns.append('t-' + str(i))
dataframe.columns = columns
pyplot.figure(1)
for i in range(1,(lags + 1)):
	ax = pyplot.subplot(240 + i)
	ax.set_title('t vs t-' + str(i))
	pyplot.scatter(x=dataframe['t'].values, y=dataframe['t-'+str(i)].values)
pyplot.show()
```
