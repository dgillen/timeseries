## Load Data
The ‘parse_dates’ parameter in the read_csv function is used to convert the ‘Date’ column to the DatetimeIndex format.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("../../data/stock_data.csv", parse_dates=True, index_col = "Date")
df.head()
```
